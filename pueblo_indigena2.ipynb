## **<font color='#3498DB'>Cargue del dataframe</font>**

```python
url = "https://raw.githubusercontent.com/Omaira85/BASES-DE-DATOS-SEG-SEMESTRE/main/Pueblo%20indigena2.csv"
puebloindigena2 = pd.read_csv(url, sep=';', encoding = 'unicode_escape')
puebloindigena2
```

## **<font color='#3498DB'>Limpieza</font>**

```python
import unicodedata

def strip_accents(s):
  if (type(s) == str):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')
  return s




for column in puebloindigena2.keys():
  puebloindigena2[column] = puebloindigena2[column].map(strip_accents)

puebloindigena2
```

### **Cambio nombre de columnas**

```python
puebloindigena2.columns = ['CODIGO', 'DEPARTAMENTO','CODIGO_1', 'MUNICIPIO',  'POBLACION_TOTAL', 'PDP_MAYA','PDP_GARIFUNA','PDP_XINCA', 'PDP_AFRODESCENDIENTE/CREOLE/AFROMESTIZO', 'PDP_LADINA(O)','PDP_EXTRANJERA(O)' ]
```

### **Se remplazan espacios en blanco por ningún valor**

```python
def remover_espacio(cadena):
  return cadena.replace(" ", "", regex = True)
```

```python
puebloindigena2.iloc[:, 4:11] = remover_espacio(puebloindigena2.iloc[:, 4:11])
```

```python
puebloindigena2
```

```python
def remover_caracteres(cadena):
  return cadena.replace("-", "0", regex = True)
```

```python
puebloindigena2.iloc[:, 4:11] = remover_caracteres(puebloindigena2.iloc[:, 4:11])
```

```python
puebloindigena2
```

### **Se cambian tipos de datos de object a int64**

```python
puebloindigena2 = puebloindigena2.astype({'POBLACION_TOTAL': 'int64', 'PDP_MAYA': 'int64', 'PDP_GARIFUNA': 'int64', 'PDP_XINCA': 'int64', 'PDP_AFRODESCENDIENTE/CREOLE/AFROMESTIZO': 'int64', 'PDP_LADINA(O)': 'int64', 'PDP_EXTRANJERA(O)': 'int64' })
```
