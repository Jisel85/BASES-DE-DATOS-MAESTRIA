Eliminación de tildes

```python
import unicodedata

def strip_accents(s):
  if (type(s) == str):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')
  return s




for column in df.keys():
  df[column] = df[column].map(strip_accents)

df
```

*Para uso en modelos*: Convertir las variables en categorías para facilitar su inclusión 

```python
df['nombre_departamento'] = df.nombre_departamento.astype('category')
df['nombre_municipio'] = df.nombre_municipio.astype('category')
df['nombre_lugar_poblado'] = df.nombre_lugar_poblado.astype('category')
df['sexo'] = df.sexo.astype('category')
df['estado_conyugal'] = df.estado_conyugal.astype('category')
df['discapacidad_permanente'] = df.discapacidad_permanente.astype('category')
df['sabe_leer_escribir'] = df.sabe_leer_escribir.astype('category')
df['asistencia_centro_educativo'] = df.asistencia_centro_educativo.astype('category')
df['razon_dejo_asistir'] = df.razon_dejo_asistir.astype('category')
df['nivel_inscrito'] = df.nivel_inscrito.astype('category')
df['nivel_aprobado'] = df.nivel_aprobado.astype('category')
df['categoria_ocupacional'] = df.categoria_ocupacional.astype('category')
df['grupo_etnico'] = df.grupo_etnico.astype('category')
df['es_jefe_familia'] = df.es_jefe_familia.astype('category')
```
